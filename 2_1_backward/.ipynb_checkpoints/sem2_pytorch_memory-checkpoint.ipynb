### Создание матриц


Класс `torch.Tensor` имеет несколько особенностей:
1. Есть специальные атрибуты: `a.T, a.H`. Они возвращают транспонированную и эрмитову матрицы, соответственно. Причем без лишнего копирования.
2. Поддерживает только `float64, float32, float16, bfloat16, int64, int32, int16, int8, uint8, bool, а также квантованные типы`
3. Можно создать новый тензор `y` с помощью уже существующего `x`: `y = torch.empty_like(x)`. Больше вариантов: [creation ops](https://pytorch.org/docs/stable/torch.html#tensor-creation-ops)
4. Ну а остальное можете прочитать в [документации к классу](https://pytorch.org/docs/stable/tensors.html#tensor-class-reference)

Хоть Pytorch выглядит python-friendly, все его вычислительно сложные операции имплементированы на C++ и CUDA.

Внутри Pytorch использует ATen - библиотеку, реализующую классы тензоров и простые тензорные операции. Для работы с градиентами используется надстройка над ATen - Autograd

### Расположение тензора в памяти

Тензоры в Pytorch бывают:
1. **Contiguous**: элементы в тензоре лежат без разрывов, в том же порядке, что и указано в размерностях.
2. **Non-Contiguous**: элементы лежат не подряд, в них есть разрывы.

А также есть частные случаи для тензоров изображений:
1. **ChannelFirst**: тензор лежит в памяти как [С,H,W], [N,C,H,W], [T,C,H,W] или [N,T,C,H,W]. Это стандартный формат для Pytorch, он contiguous.
2. **ChannelsLast**: тензор лежит в памяти как [H,W,C], [N,H,W,C], [T,H,W,C] или [N,T,H,W,C]. С точки зрения Pytorch это формат для специальных операций


Операции в Pytorch ожидают на вход **ChannelsFirst**-тензоры, но могут на выходе формировать любого вида

Порядок хранения элементов тензора называют **layout**. Он показывает, как класс Tensor логически интерпретирует физическую память.


Представление **ChannelsLast** может дать прирост производительности. Когда элементы в каналах для каждого пикселя будут лежать в памяти локально, у устройств не будет дополнительных расходов на подготовку сложных индексов и извлечения данных по этим индексам - это очень удобно, например, для операции свертки, ведь она идет по всем каналам и только по части пространственных координат.

Так [тензорные ядра](https://developer.download.nvidia.com/video/gputechconf/gtc/2019/presentation/s9926-tensor-core-performance-the-ultimate-guide.pdf) работают быстрей c **ChannelsLast**.
Также некоторые библиотеки (например, [XNNPACK](https://github.com/google/XNNPACK)), которые Pytorch использует в качестве бэкенда, поддерживают **ChannelsLast** и для них не нужно будет менять layout тензора.

Но **ChannelsFirst** быстрее работают в batchnorm-слоях, а также везде, где есть broadcast по H и W.

````python
# давайте посмотрим на примере, как выглядят разные форматы хранения тензоров
# создадим тензор [N,C,H,W]
a_contiguous = torch.arange(16).view(1,2,2,4)
# функция torch.Tensor.storage() вернет тензор в том виде, в каком он лежит в памяти
# то есть, 1d-представление тензора
print(a_contiguous.storage())
````

````python
# посмотрим на логическое отображение тензора
print(a_contiguous)
````

````python
# а теперь поменяем формат на channels_last
a_channels_last = a_contiguous.to(memory_format=torch.channels_last)
print(a_channels_last.storage())
````

````python
# логическое отображение
print(a_channels_last)
````

````python
print(a_contiguous.is_contiguous())
print(a_channels_last.is_contiguous())
# укажем pytorch способ интерпретации storage
print(a_channels_last.is_contiguous(memory_format=torch.channels_last))
````

**Storage** - обертка над указателем - ничего не говорит о том, как интерпретировать данные. Таким образом, можно сформировать несколько тензоров (несколько view) над одим storage без копирования с помощью различных strides

**Stride** - это расстояние шага от одного элемента к другому для конкретной размерности. Для каждой размерности соответствующий элемент страйда показывает,
какой длины должен быть шаг, чтобы перейти к следующему элементу массива в 1d-представлении

image.png

````python
# давайте посмотрим, как изменяются страйды
# при переходе от одного формата к другому
a_contiguous = torch.rand(128, 1024, 7, 4)
a_channels_last = a_contiguous.to(memory_format=torch.channels_last)
# как видно на этом примере, элементы лежат в row-major order
print("a_contiguous stride:", a_contiguous.stride())
print("a_channels_last stride:", a_channels_last.stride())

# причем shape этих тензоров один и тот же!
print("a_contiguous shape:", a_contiguous.shape)
print("a_channels_last shape:", a_channels_last.shape)
````

````python
# связь storage и stride
a_strides = a_contiguous.stride()
idx = (0, 1, 2, 3) # будем брать этот элемент из тензора
loc = sum([i*s for i,s in zip(idx, a_strides)]) # посчитаем его индекс в 1d-представлении
print("Это один и тот же элемент?", (a_contiguous.storage()[loc] == a_contiguous[idx]).item())
````

````python
# вот так страйд позволяет избежать копирования
# можно создать новый тензор, view, который будет указывать на старый
# но у которго будет другой страйд
a_strided = a_contiguous[::2, ::4, ::7]
print("Strides of slice", a_strided.stride())
print("Is a_strided contiguous: ", a_strided.is_contiguous())
````

````python
# чтобы проверить, что копирования не произошло, можно in-place изменить a_strided
# и посмотреть, что произошло с a_contiguous
# a_strided *= 1000
# print(a_contiguous)
````

````python
# аналогично expand не требует дополнительной памяти
# то есть, здесь нет копирования
a_contiguous = torch.arange(3).view(1, 3)
y = a_contiguous.expand(2, 3)
z = a_contiguous.repeat(2, 1)

# кажется, что результат одинаковый
print("Expand:\n", y)
print("Repeat:\n", z)

# но объем занимаемой памяти совсем разный!
print("Y:", len(y.storage()) * y.element_size())
print("Z:", len(z.storage()) * z.element_size())

print("Is expanded tensor contiguous", y.is_contiguous())
print("Is repeated tensor contiguous", z.is_contiguous())
````

Есть несколько способов получить из contiguous тензора non-contiguous:
1. Транспонирование и permute тензора
2. Slicing (как видно на примере выше)
3. View и reshape
4. Expand (мы разобрали и этот случай)

Иначе говоря, эти функции переставляют страйды, чтобы избежать лишнего копирования.  
Но тогда теряется локальность данных и многие операции могут выполняться медленнее. Могут - но только если вы используете низкоуровневые операции, потому что Torch сам следит за тем, когда из non-contiguous нужно сделать contiguous для более оптимального исполнения.  
Однако если вы пишете свой CUDA или CPU kernel, то вам придется передовать входные данные по указателю (с помощью torch.Tensor.data_ptr). В этом случае придется или учитывать страйды, или сделать тензор contiguous




### Матричные операции и broadcast

Два тензора поддерживают broadcast, если:

1. У каждого тензора есть как минимум одна размерность
2. Значения размерностей двух тензоров должны быть равны, или одна из них должна быть единицей, или одна из них отсутствует. Сравнение значений начинается с последней размерности

Размер тензоров после broadcast'а устанавливается так:
1. Если размерность отсутствовала, на ее место ставится 1
2. Далее все размерности выбираются как максимальные значения в соответствующей паре размерностей тензоров


````python
a = torch.randn(4, 5)
b = torch.randn(5, 4)
# матричное умножение
c = a @ b
print('Результат: \n', c)
print('Размер матрицы: ', c.shape)
````

````python
# поэлементное умножение
b = torch.transpose(b, 0, 1)
c = a * b
print('Результат: \n', c)
print('Размер матрицы: ', c.shape)
````

````python
# broadcast
# a[4,5] + 2 -> a[4,5] + 2[4,5]
# Помните, может, weight decay в линейной регрессии?
# Там тоже своеобразный broadcast, только не с матрицей единиц, а с единичной матрицей :)
c = a + 2
print('Матрица а: \n', a)
print('Результат: \n', a + 2)
````

````python
# a[4,5] + b[5] -> a[4,5] + b[1,5] -> a[4,5] + b[4,5]
# broadcast по столбцам
b = torch.arange(5)
print(b.shape)
c = a + b
c
````

````python
# сделаем, чтобы ячейка не выбрасывала ошибку
%%script false --no-raise-error

# broadcast по строкам
# так не бродкаст не заработает
# ведь размеры тензоров не совпадают по 0 размерности (если считать с конца)
# a[4,5] + b[4]

b = torch.arange(4)
c = a + b
````

````python
b = torch.arange(4)
b = b.unsqueeze(1)
# broadcast по строкам
# b[4] -> unsqueeze(1) -> b[4, 1]
# a[4,5] + b[4,1] -> a[4,5] + b[4,5]
c = a + b
c
````

````python
# broadcast для обеих матриц (outer product)
a = torch.arange(5, 10).reshape(1,5)
b = torch.arange(5).reshape(5,1)
c = a * b
print(c)
print(c.shape)
````

````python
# broadcast бывает разный

# считаем размерности с конца
# 0 dimension: обе размерности равны 1
# 1 dimension: у b размерность 1
# 2 dimension: у a размерность 1
# 3 dimension: обе размерности равны 5
# 4 dimension: у b отсутствует
a = torch.randn(3, 5, 1, 6, 1)
b = torch.randn(   5, 4, 1, 1)
c = a / b
print(c.shape)
````

````python
# batched matrix multiplication (BMM)
# здесь перемножаем матрицы для каждого элемента из батча независимо
a = torch.randn(20, 3, 16, 512)
b = torch.randn(20, 3, 512, 16)
c = a @ b
print(c.shape)
````

Для корректного матричного умножения тензоров высоких размерностей стоит прочитать документацию:

### Автоматическое дифференцирование

У каждого тензора в Pytorch есть флаг `requires_grad`, который отвечает за автоматическое вычисление градиентов:

1. Создать переменную: `a = torch.tensor(..., requires_grad=True)`

2. Определить какую-нибудь дифференцируемую функцию `loss = whatever(a)`

3. Запросить обратный проход `loss.backward()`

4. Градиенты будут доступны в `a.grads`


Есть два важных отличия Pytorch от Theano/TF:

1. Функцию ошибки можно изменять динамически, например на каждом минибатче.

2. После вычисления `.backward()` градиенты сохраняются в `.grad` каждой задействованной переменной, при повторных вызовах градиенты суммируются. Это позволяет использовать несколько функций ошибок или виртуально увеличивать batch_size. Поэтому, после каждого шага оптимизатора градиенты стоит обнулять (или [присваивать им None](https://pytorch.org/docs/stable/generated/torch.optim.Optimizer.zero_grad.html) для экономии памяти, как это делает `torch.zero_grad`).


**Leaf vs Non-leaf Variable:**

Градиенты будут сохранены и доступны для использования только для `leaf tensor`.
Такое поведение по-умолчанию сделано ради экономии памяти. Все тензоры с флагом `requires_grad = False` считаются `leaf tensors` по умолчанию.


Обратите внимание, что вычисление градиентов работает только для тензоров с вещественным типом данных:

````python
x = torch.tensor([1., 2., 3., 4.], requires_grad=True) # leaf tensor
y = x + 1  # not a leaf variable
z = (y ** 2 + x).sum()
z.backward()

print(x.grad)
````

Чтобы выставить флаг `requires_grad=False` и выключить автоматическое вычисление градиентов для нескольких тензоров, можно использовать `with torch.no_grad()`, `detach` или `torch.inference_mode` (подробнее о различиях между ними [здесь](https://pytorch.org/docs/stable/notes/autograd.html#locally-disabling-gradient-computation)):

````python
x = torch.tensor([1.], requires_grad=True)
y = x**2
print(x.requires_grad)
print(y.requires_grad)

with torch.no_grad():
  z = torch.exp(x)
print(z.requires_grad)

with torch.inference_mode():
  # может быть быстрее, чем torch.no_grad
  z = torch.exp(x)
print(z.requires_grad)

# detach from the graph
w = torch.log(x).detach()
print(w.requires_grad)
````

### Создание вычислительного графа и выделение памяти в cuda

Пример вычислительного графа для функции `f = y(x1, x2, x3, w)`:


```
r1 = x3 * (x1 / x2 + w[1] / w[0])
r2 = x1 * x2 * x3 / (w[0] + w[1])
r3 = (w[0] * w[1]) ** 3.0
y = (r1 + r2 + r3) / (x1 * r1 + r2 + r3)
```



image.png

Данный граф визуализирован с помщью [netron.app](https://netron.app)
Для того, чтобы вам самим получить такой рисунок, нужно:
1. Обернуть функцию в класс, унаследованный от `nn.Module`, и переопределить метод `forward`, поскольку именно он отвечает за построение вычислительного графа:
```
class Graph(nn.Module):
    
      def __init__(self):
          super().__init__()
          self.w = nn.Parameter(torch.randn(2, dtype=torch.float32))
      
      def forward(self, x1, x2, x3):
          r1 = x3 * (x1 / x2 + self.w[1] / self.w[0])
          r2 = x1 * x2 * x3 / (self.w[0] + self.w[1])
          r3 = (self.w[0] * self.w[1]) ** 3.0
          return (r1 + r2 + r3) / (x1 * r1 + r2 + r3)
```
`nn.Parameter` - субкласс `torch.Tensor`, показывающий, что данный атрибут - параметр описываемого нами моделя. По умолчанию имеет `requires_grad=True`
2. Экспортировать Pytorch-модeль в формат onnx:
```
  model = Graph()

  torch.onnx.export(
      model,
      (x1, x2, x3),
      'graph.onnx',
      opset_version=13,
      export_params=True,
      do_constant_folding=False,
      input_names=['x1', 'x2', 'x3'],
      output_names=['y'],
  )
```

[ONNX](https://github.com/onnx/onnx) - формат вычислительного графа. Он позволяет абстрагироваться от представлений вычислительных графов, связанных с конкретными реализациями (скажем, форматов .pt или .ckpt), и перейти непосредственно к работе с узлами графа.
Также ONNX позволяет проводить оптимизацию графа, выводить shape и type для промежуточных тензоров и поддерживается множеством различных фреймворков для инференса: OnnxRuntime, TensorRT и OpenVino

````python
import time
import torch

def cuda_memory():
    torch.cuda.synchronize()
    torch.cuda.empty_cache()
    time.sleep(0.5)
    print(torch.cuda.memory_allocated(0))
````

````python
cuda_memory()
````

````python
x = torch.ones(500, 500, dtype=torch.float32, requires_grad=True, device='cuda:0')
cuda_memory()
````

PyTorch хранит в памяти только те тензоры, которые необходимы для вычисления градиентов. Если же при `forward` или `backward` появляются неиспользуемые переменные, PyTorch удаляет их из графа.

Еще один подход для экономии памяти применяется и методе [*activation checkpointing*](https://pytorch.org/docs/stable/checkpoint.html): мы храним только ту часть тензоров, появляющихся при forward'е вычислительного графа, которая необходима для вычисления градиентов на текущем шаге backprop. Когда мы переходем к следующему шагу, считаем `forward` еще раз.
Этот метод затратен по вычислительным ресурсам, но позволяет экономить память и считатать даже очень большие сети в условиях ограниченной памяти.

```
We show that for a 30% increase in FLOPs, checkpointing can reduce the memory required for the activations by 5.8x on WideResNet
and 5.7x on DC-Transformer.
```


Давайте посмотрим, как работает выделение и освобождение памяти на конкретном примере:



---


Расмотрим вычислительный граф следующей функции: $$a(x)= sum((x + x + x + x)^2)$$
Давайте попробуем понять, сколько памяти Pytorch должен будет выделить

Для этого распишем forward вычислительного графа:
```
a1 = x + x
a2 = a1 + x
```

Больше не нужен `a1`, поскольку его градиент можно посчитать без значений `а1`
поэтому удалим его
```
a1 = None
a3 = a2 + x
a2 = None
a4 = a3 ** 2
```

Мы не можем удалить `а3`, он понадобится при вычислениях. А вот `а4` - можем
```
a = a4.sum()
a4 = None
```

### Трейсинг графа и компиляция

Нейросеть - такая же программа, как и все остальные.
Она так же может быть представлена в виде графа - а значит, может быть скомпилирована и оптимизирована под конкретные устройства.

> Компилятор для нейронных сетей называется **тензорным компилятором** (более подробно можете прочитать [здесь](https://habr.com/ru/companies/yadro/articles/782208/) и [здесь](https://github.com/merrymercy/awesome-tensor-compilers))

Главные оптимизации, которые выполняет компилятор:
1. **Kernel fusion** - объединяет kernel'ы различных операций в один kernel там, где это возможно. Это помогает избравиться от лишних операций чтения-записи и накладных расходов на создание kernel'ов.
2. **Управление памятью** - сколько памяти необходимо аллоцировать для модели, чтобы не переаллоцировать, поскольку это дорогая операция
3. **Оптимизации под архитектуру**: оптимизация под вычислительные ядра, количество и устройство памяти, поддерживаемые операции



**Этапы компиляции** (очень примерно):
1. Получение вычислительного графа помощью генерации *Intermediate Representation* (IR), языка представления Abstract Syntax Tree, генерируемого фронтэндом компилятора. С помощью синтаксиса IR выполняются оптимизации. Таких языков много, поэтому компиляторы поддерживают не какой-то один IR, а *семейство диалектов*.
> В терминах IR оптимизация кода сводится к однотипным оптимизациям: вставке узлов (то есть, инструкций), удалении, замены одной инструкции на другую.
2. Алгебраические оптимизации (сплавление линейных операций, перестановки нод для возможных kernel fusion, dead node ellimenation, проведение tiling). Такие оптимизации не зависят от конечного устройства (то есть, они *device-independent*), иными словами, если их сделать, то граф будет рабочим на любом устройстве.
> Все оптимизации в IR основаны на pattern matching: если в графе найдена определенная комбинация, то она заменяется на другую определенную комбинацию. Например, свертка заменяется на две свертки и редукцию
3. *Device-specific* оптимизации (как и где должен проводиться broadcast, замена высокоабстрактных операций на более конкретные): оптимизации, которые зависят строго от архитектуры.
5. Генерация Low Level IR. Это может быть сразу ассемблер.
6. *Device-specific* оптимизации. Расставление порядка вычисления нод (device placement and scheduling)
9. Генерация машинного кода (code emmiting или иначе называется code generation)
10. Runtime: интерфейс, который обеспечивает взаимодествие пользователя и устройства.

При переходе от одних этапов компиляции к другим IR становится все более низкоуровневым - так, чтобы на нем можно было более удобно и точно описывать граф

Чем это отличется от обычного выполнения pytorch-модели?
1. Обычный тип исполнения графа в Pytorch называется **eager mode**. Граф в нем динамический, поэтому каждая операция в runtime вызывает нужный kernel. Из-за этого нельзя выполнить kernel fusion, алгебраические и остальные оптимизации. Но зато графы в eager mode легко отлаживать.
2. Также из-за времени, затрачиваемого на запуск кернела, возникает проблема низкой утилизации GPU. Для ее решения использовали, например, [CUDA Graphs](https://pytorch.org/blog/accelerating-pytorch-with-cuda-graphs/) - библиотеку, позволяющую сплавлять некоторые кернелы и тем самым уменьшать CPU overhead (то есть, время, когда работает только CPU, а GPU простаивает)

> Также в Pytorch поддерживается Just In Time (JIT) компиляция - кернелы компилируются при запуске программы или при вызове кернела.

В Pytorch компиляция поддерживается через **torch.dynamo** в несколько этапов:
1. Анализ Python байт-кода и составление вычислительного графа
2. Динамическая модификация байт-кода через компиляцию вычислительного графа под конкретное устройство
3. Проверка корректности среды и входных данных через специальные условия - gurads. Если одно из них не выполнится, граф отправится на перекомпиляцию
4. Исполнение модифицированного байт-кода

````python
# Теперь давайте посмотрим, как в работает pytorch в режиме компиляции

import os
# выставим отладочную переменную
os.environ["TORCH_COMPILE_DEBUG"] = "1"

# определим функцию для трейсинга
def sin_fn(x):
  if x.sum() > 0:
    b = torch.sin(x + 1) + torch.sin(x + 2) + torch.sin(x + 3) + torch.sin(x + 4)
  else:
    b = 10 * x
  return b

# здесь строится точка входа
# inductor - бэкенд компилятора, то есть, набор kernel'ов, написанных для конкретных устройств
# на CUDA, Triton для GPU или на С++ OpenMP для CPU
new_fn = torch.compile(sin_fn, backend="inductor")
x_zeros = torch.zeros_like(x)
````

Программу можно описать не только через AST, но и через **граф потока данных** (data flow) - то есть, то, как данные изменяются и какими операторами в теле программы.

Формально граф потока данных - ориентированный граф, вершины которого являются инструкциями, а ребра - переходы данных между инструкциями. При слиянии ребер необходимо решить, откуда выбрать значение - это решает **поток управления** (control flow).

Именно через представление graph data flow современные компиляторы (такие как LLVM) оптимизируют и обрабатывают код.

В torch dynamo потоки управления поддерживаются с помощью компиляции различных подграфов

Явно получить граф без компиляции также можно средствами Pytorch с помощью модуля torch.fx, предоставляющего инструменты для **трейсинга графа**

Метод `fx.symbolic_trace` покажет семантику модуля. Он пропускает *прокси-тензор* через функцию, отправленную в него, и записывает все операции, которые воздействуют на этот тензор. Таким образом он получает data flow graph.

Граф в `fx.symbolic_trace` будет отличаться от предыдущих графов, сгенерированных через `torch.compile`: и полученного в результате анализа байт-кода, и полученного после компиляции. Поскольку он смотрит не на байт-код, а на операции pytorch

````python
from torch import fx

# определим функцию для трейсинга
# без compiler control flow!
# потому что fx.trace зависит от входных данных
def sin_fn(x):
  b = torch.sin(x + 1) + torch.sin(x + 2) + torch.sin(x + 3) + torch.sin(x + 4)
  return b

# прокси-тензор может быть случайным или фиксированным
# с помощью фиксирования тензора мы можем обойти все подграфы, возникающие при условиях
symbolic_traced : torch.fx.GraphModule = fx.symbolic_trace(sin_fn)
````

````python
print("High-level intermediate representation (IR) - Graph representation")
print(symbolic_traced.graph)

# Code generation - valid Python code
print(symbolic_traced.code)
````

````python
# еще один способ вывода графа
print(symbolic_traced.graph.print_tabular())
````

Зачем нам нужно самим получать граф?
1. Для получения промежуточного представления. Это представление биективно отображает код и модель.
2. Можно добавлять дополнительные слои, которые повлияют на процесс компиляции: например, Quant и Dequant ноды для построения схемы квантования. Или заменять некоторые слои на кастомные, но котрые знакомы компилятору
3. Можно написать конвертер из произвольного формата в torch: например, так устроен [onnx2torch](https://github.com/ENOT-AutoDL/onnx2torch/tree/main)

Больше про устройство графа в pytorch и его компиляции вы можете прочитать [вот тут](https://blog.christianperone.com/2023/12/pytorch-2-internals-talk/)
